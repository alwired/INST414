```python
import pandas as pd

df = pd.read_csv('data/songs.csv')
df.head()
df.shape
```

```python
df = df.drop_duplicates(subset=['track_id'])
df = df.dropna()

names = df[['track_id', 'artist_name', 'track_name']]
vals = df.drop(['genre', 'artist_name', 'track_name', 'key', 'mode', 'time_signature'], axis = 1)
track_genres = df[['track_id', 'genre']]
modes = df[['track_id', 'mode']]
modes.loc[modes['mode'] == 'Major', 'mode'] = 1
modes.loc[modes['mode'] == 'Minor', 'mode'] = 0

names.set_index('track_id', inplace=True)
vals.set_index('track_id', inplace=True)
track_genres.set_index('track_id', inplace=True)
modes.set_index('track_id', inplace=True)


vals.head()
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(vals)
scaled_vals = vals.copy()
scaled_vals[:] = scaler.transform(vals)
scaled_vals
```
